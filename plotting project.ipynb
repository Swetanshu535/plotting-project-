```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df =pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Disney_princess.csv")
print(df)
```

Bar Chart â€“ Number of Songs per Princess
Question:
Plot a bar chart showing the total number of songs each princess has across all her appearances.


```python
song_total=df.groupby("PrincessName")["NumberOfSongs"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
song_total.plot(kind='bar',color='red')
plt.title("bar chart for total number of songs")
plt.xlabel("PrincessName")
plt.ylabel("NumberOfSongs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Question:
Plot a bar chart showing the total HasDuet and MovieRuntimeMinutes 

```python
movie_runtime=df.groupby("HasDuet")["MovieRuntimeMinutes"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
movie_runtime.plot(kind="bar",color="purple")
plt.title("Movie run time in minutes and has duet or not")
plt.xlabel("HasDuet")
plt.ylabel("MovieRuntimeMinutes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Create a bar chart that shows the distribution of different HairColor values among all the princesses.

```python
haircolor=df.groupby("HairColor")["NumberOfSongs"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
haircolor.plot(kind="bar",color="orange")
plt.title("Haircolor and NumberOfSongs")
plt.xlabel("HairColor")
plt.ylabel("NumberOfSongs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
